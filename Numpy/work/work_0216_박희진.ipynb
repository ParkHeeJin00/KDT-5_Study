```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

### 1번

```python
df = pd.read_csv('../data/CARD_SUBWAY_MONTH_202312.csv', usecols=[0,2,3,4])
```

```python
df = df.set_index(df['역명'])
```

```python
df['이용객수'] = (df['승차총승객수'] + df['하차총승객수']) /2
```

```python
for i in df['역명'].unique():
    print(f'{i}:{df.loc[i,"이용객수"].mean()}')
```

```python
for i in df['역명'].unique():
    print(f'{i}:{df.loc[i,"이용객수"].std()}')
```

```python
np.max(df['이용객수'])
```

```python
np.min(df['이용객수'])
```

```python
df['이용객수'].idxmax()
```

```python
df['이용객수'].idxmin()
```

```python
plt.hist(df['이용객수'], bins = 50)
plt.title('[서울 지하철 역별 이용객수의 분포]')
plt.xlabel('이용객수')
plt.ylabel('해당하는 지하철 역 수')
plt.show()
```

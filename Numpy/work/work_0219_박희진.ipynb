```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
df = pd.read_csv('../data/iris.csv')
df.set_index('variety', inplace=True)
df
```

## 1번

```python
n = 1
for i in df.columns[:-1]:
    for j in df.columns[n:]:
        print(f'{i}와 {j}의 공분산 : {np.cov(df[i],df[j], ddof = 0)[0,1]}')
    n += 1
```

```python
df.cov()
```

## 2번

```python
n = 1
for i in df.columns[:-1]:
    for j in df.columns[n:]:
        print(f'{i}와 {j}의 상관계수 : {np.corrcoef(df[i],df[j], ddof = 0)[0,1]}')
    n += 1
```

```python
df.corr()
```

## 3번

```python
df.columns
```

```python
i = 1
for c in df.columns:
    plt.subplot(2,2,i)    
    plt.hist(df[c], color = 'orange')
    i += 1
    plt.title(f'{c}')
    plt.xlabel('길이')
    plt.ylabel('개수')
plt.tight_layout()
plt.suptitle('[특성별 히스토그램]')
```

## 4번

```python
i = 1
for c in df.columns:
    plt.subplot(3,4,i)
    plt.boxplot(df.loc['Setosa',c])
    i +=1
    plt.title(f'{c}', size = 9)
plt.suptitle('[Setosa 특성에 대한 boxplot]')
```

```python
i = 1
for c in df.columns:
    plt.subplot(3,4,i)
    plt.boxplot(df.loc['Versicolor',c])
    i +=1
    plt.title(f'{c}', size = 9)
plt.suptitle('[Versicolor 특성에 대한 boxplot]')
```

```python
i = 1
for c in df.columns:
    plt.subplot(3,4,i)
    plt.boxplot(df.loc['Virginica',c])
    i +=1
    plt.title(f'{c}', size = 9)
plt.suptitle('[Virginica 특성에 대한 boxplot]')
```

```python
plt.figure(figsize=(5,10))
n = 1
for i in df.index.unique():
    plt.subplot(3,1,n)
    plt.boxplot([df.loc[i,'sepal.length'], df.loc[i,'sepal.width'], df.loc[i,'petal.length'], df.loc[i,'petal.width']])
    plt.title(f'{i} 특성에 대한 boxplot')
    n +=1
    plt.xticks(np.arange(1,5),['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
plt.tight_layout()
```

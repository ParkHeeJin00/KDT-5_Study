```python
### 모듈로딩
from PIL import Image
from torchvision import transforms
from torchvision.datasets import ImageFolder
```

```python
### 이미지 데이터 로딩
file = '../data/img/7e0133768625340913c220136a31fef2.jpg'
```

```python
img = Image.open(file)

# img 속성 알아보기
img.format, img.height, img.width, img.mode, type(img)
```

```python
# 사용법 => (1) 인스턴스 생성
to_tensor = transforms.ToTensor()

### (2) 인스턴스 변수 이용
imgTS = to_tensor(img)

### (3) 변환된 이미지 텐서 확인
imgTS.shape, imgTS.ndim, imgTS[0]
```

```python
### openCV
import cv2

img1 = cv2.imread(file)  # ndarray 반환 - BGR로 읽어옴
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)  # 색상 채널 위치 RGB로 변환
img2.shape, img2.ndim, type(img2)

### 텐서화
imgTS2 = to_tensor(img1)
imgTS3 = to_tensor(img2)

imgTS2.shape, imgTS2.ndim, imgTS2[0], imgTS3[0]
```

```python
preprocessing = transforms.Compose([

    transforms.Resize((28,28)),  # 사이즈가 작은 쪽에 맞춰서 조절
    transforms.ToTensor()
]) # 정규화는 무조건 마지막에 !!
```

```python
preprocessing(img)
```

### 이미지 데이터셋 <hr>
- torchvision.datasets 클래스 사용
    * 이미지 데이터 라벨링
    * 이미지 데이터 전처리

```python
from torchvision.datasets import ImageFolder
```

```python
img_root = '../data/img'
imgDS = ImageFolder(root = img_root, transform=preprocessing)
```

```python
# 폴더명이 분류 클래스
# 폴더 이름  /     이름, 라벨        /  이미지 주소 
imgDS.classes, imgDS.class_to_idx, imgDS.imgs
```

```python
from torch.utils.data import DataLoader
```

```python
imgDL = DataLoader(imgDS)
```

```python
for (feature, label) in imgDL:
    print(feature.shape, label.shape)
    break
```

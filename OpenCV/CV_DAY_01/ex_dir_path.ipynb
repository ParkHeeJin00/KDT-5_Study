```python
### 모듈 로딩
import os
```

```python
# 특정 폴더 내의 모든 데이터 리스트 가져오기
my_folder = '../data/'

# 해당 폴더 또는 파일이 존재하는지 검사 ===> 결과 :True, False
os.path.exists(my_folder)
```

```python
# 폴더 내 모든 파일, 폴더 리스트 반환
datalist = os.listdir(my_folder)  # 파일과 폴더 이름 모두 반환

for dataname in datalist:    
    # 폴더 여부 체크 => os.path.isdir(경로)
    print(dataname, os.path.isdir(my_folder+dataname), os.path.isfile(my_folder+dataname))
```

```python
#
for dataname in datalist:
    # 폴더 여부 체크 후 폴더일 경우
    if os.path.isdir(my_folder+dataname):
        folder_path = my_folder+dataname
        print(f'[{dataname}]')
        for file in os.listdir(folder_path):
            if os.path.isfile(folder_path+'/'+file):
                print('----',file)
    # 폴더가 아닌 경우
    else:
        print(dataname)
    # 확장자 검사는 split으로 !!
```

```python
# 원하는 데이터 파일 즉 확장자를 지정해서 해당 파일만 가져오기
def getFiles(dir,extension):
    for dataname in os.listdir(dir):
        # 폴더 여부 체크 후 폴더일 경우
        if os.path.isdir(my_folder+dataname):
            folder_path = my_folder+dataname
            print(f'[{dataname}]')
            for file in os.listdir(folder_path):
                if file.endswith(extension):      # os.path.splitext(file)[-1] -> 확장자만 반환 가능
                    print('----',file)
        # 폴더가 아닌 경우
        else:
            if file.endswith(extension):
                print(dataname)
```

```python
getFiles(my_folder,'csv')
```

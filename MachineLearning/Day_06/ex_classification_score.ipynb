### 분류 모델 성능지표
- 모델 구현 후 성능 평가 방식
- 정확도(Accuracy) : 불균형 데이터에 부적합
- 보완 평가 방식 : 정밀도, 재현율, F1점수, 혼돈행렬 테이블

(1) 임시 데이터 <hr>

```python
y_true = [0,1,1,0]
y_pre = [1,1,1,0]
```

(2) 분류 평가지표 <hr>

```python
## 분류 정확도 => accuracy : 불균형 데이터의 경우 모델 평가 지표로 사용 불가
from sklearn.metrics import accuracy_score

accuracy_score(y_true, y_pre)
```

```python
## 정밀도 => precision : 예측 양성율
from sklearn.metrics import precision_score

precision_score(y_true,y_pre)
# 3개를 p라고 했는데 2개 맞춤
```

```python
## 재현율 => recall : 양성율
from sklearn.metrics import recall_score

recall_score(y_true,y_pre)
# p가 총 두개인데 다 맞춤
```

```python
# f-1 점수 => f1-score : 정밀도와 재현율 상호보완 조화평균
from sklearn.metrics import f1_score

f1_score(y_true,y_pre)
```

```python
# 혼돈 행렬값 출력
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

confusion_matrix(y_true,y_pre)
```

```python
ret_matrix = confusion_matrix(y_true,y_pre)
tn,fp,fn,tp = confusion_matrix(y_true,y_pre).ravel()
#  confusion_matrix(y_true,y_pre).reshape(-1)도 가능
tn,fp,fn,tp
```

```python
# 정확도, 정밀도, 재현율, f1-score 계산

# 정확도
ac = (tn+tp) / (tn+fp+fn+tp)
print(f'정확도 : {ac}')
# 정밀도
sp = tp/(tp+fp)
print(f'정밀도 : {sp}')
# 재현율
rc = tp / (tp+fn)
print(f'재현율 : {rc}')
# f1-score
f1 = 2 * sp * rc / (sp+rc)
print(f'F1-Score : {f1}')
```

```python
import matplotlib.pyplot as plt

cls = ConfusionMatrixDisplay(ret_matrix)
cls.plot(cmap = 'bone')
plt.show()
```

### 표본 추출/샘플링
- 전체 데이터에서 일부 데이터를 추출
- pandas.DataFrame.sample() 메서드
    - n : 갯수 
    - frac : 비율(0.0 - 1.0)
    - n과 f 파라미터 동시에 사용 불가
- 복원추출 : 추출된 데이터를 다시 추가해서 추출하는 방식
- 비복원 추출 : 추출된 데이터를 제거 후 남은 데이터에서 추출하는 방식

```python
import pandas as pd
```

```python
# 임의의 데이터 생성
dataDF = pd.DataFrame({'kor':[10,20,30],
                       'math':[11,22,33],
                       'eng':[111,222,333]})
dataDF
```

```python
# 1개의 샘플/표본 데이터 추출 => [기본] 비복원 - 전체 데이터 수 초과 불가
# 추출 기준 축(axis)로 설정
for cnt in range(5):
    print(f'{cnt}번 : {dataDF.sample(n = 1, axis=1)}') # 행 추출(기본값) : axis = 0 // 열 추출 -> axis = 1 or axis = 'columns'
```

```python
# 1개의 샘플/표본 데이터 추출 => [설정] 복원 : replace 매개변수 True로 설정
# 추출 기준 축(axis)로 설정
for cnt in range(5):
    print(f'{cnt}번 : {dataDF.sample(n = 5, axis=1, replace=True)}')
```

```python
# 모듈로딩
import imageio.v2 as imageio           # 이미지 데이터 관련 모듈
import torch              # Pytrch 텐서 및 기본 함수 관련 모듈
```

```python
# 이미지 데이터 로딩
filename = '../img/KakaoTalk_20240106_133758930.jpg'
```

```python
img_data = imageio.imread(filename)
print(img_data.shape, img_data.ndim, img_data.dtype)
```

```python
# ndarray => tensor 변환 : [행(높이) : 열(너비) : 채널] 
img_t1 = torch.tensor(img_data)  # 복사본 생성 / 별개 : 데이터 공유 X
img_t2 = torch.from_numpy(img_data)
img_t3 = torch.as_tensor(img_data)
```

```python
print(img_t1.shape, img_t1.ndim)
print(img_t2.shape, img_t2.ndim)
print(img_t3.shape, img_t3.ndim)
```

```python
# [행(높이) : 열(너비) : 채널] ==> [채널 : 행(높이) : 열(너비)]
# 채널 : 삼색   / 흑백이미지는 채널이 1 / icon과 같이 투명도가 있는 사진은 채널이 4 (rgba) / 대부분 이미지는 채널이 3 (rgb)
# 배치사이즈는 한번에 보는 값 - 32가 보통 성능이 좋음
img_t1.permute(2, 0, 1)
img_t2.permute(2, 0, 1)
img_t3.permute(2, 0, 1)
```

```python
# 데이터 파일 => CSV 파일 : 분리자 - : , 첫번째 줄 - 컬럼명, 데이터 타입 - 실수
datafile = '../data/text/winequality-white.csv'
```

- [2-1] Pandas 사용한 데이터 로딩

```python
import pandas as pd

data = pd.read_csv(datafile, sep=';')

data = data.astype('float32') 
feature = data[data.columns[:-1]]
target = data[data.columns[-1]]

feature_tensor = torch.from_numpy(feature.values)
target_tensor = torch.from_numpy(target.values)
target_tensor.shape # 1차원
```

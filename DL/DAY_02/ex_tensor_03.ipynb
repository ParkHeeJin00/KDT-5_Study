## 행렬 곱셈
- 피쳐와 가중치 연산에 활용
- tensor.matmul()

```python
# 모듈 로딩
import torch
```

```python
data = torch.tensor([[1,2],[3,4]])
w = torch.tensor([7])
```

```python
data.shape, w.shape
```

```python
# 원소단위 곱셈 => 인덱스에 1:1 대응하도록 브로드캐스팅
data.mul(w)
```

```python
# 행렬단위 곱셈 => 행x열 : (행, 열) x (행, 열) == (행, 열) x (원소수, )
w = torch.tensor([7,8]) # 1차원 => 틀림
w2 = torch.tensor([[7],[8]]) # 2행 1열
data.matmul(w2), w2.shape
```

### 텐서 연결 & 결합

```python
# 모듈 로딩
import torch
```

```python
# 텐서 데이터 생성
torch.manual_seed(12) # 랜덤 초기값 설정
# 임의의 0. <= ~ <= 1. 실수
x = torch.rand(2,10)  # 2행 10열
# 임의의 정수 : low <= ~ <= high
y = torch.randint(0,2,(10,)) # 1차원
y2 = torch.randint(0,2,(2,10)) # 2행 10열
```

```python
print(x.shape, x.ndim,sep = '\n')
print(y.shape, y.ndim, sep = '\n')
print(y2.shape, y2.ndim, sep = '\n')
```

```python
# tensor 연결 cat([텐서1, 텐서2....])
xy1 = torch.cat([x,y2], dim = 1)
```

```python
# 집계함수는 aixs나 dim  기준이 반대
```

```python
# 행 단위로 최대값 및 최대값의 인덱스
max_value = xy1.max(dim =1)[0]
max_idx = xy1.max(dim =1)[1]

max_value, max_idx
```

```python
# 열 단위로 최대값 및 최대값의 인덱스 # dim = 0 or -1
max_value = xy1.max(dim =0)[0]
max_idx = xy1.max(dim =0)[1]

max_value, max_idx
```

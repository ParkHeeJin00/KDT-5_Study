```python
import torch
import torch.nn as nn
import torchviz
import torchinfo
```

```python
class Net(nn.Module):
    
    # 모델 구성 요소 생성 및 구조 설정
    def __init__(self):
        super(Net, self).__init__()  # super() 도 가능
        
        self.fc1 = nn.Linear(8,4)
        self.fc2 = nn.Linear(4,2)
        self.fc3 = nn.Linear(2,1)
    
    # 순방향 학습 진행 함수
    def forward(self, x):
        return self.fc3(self.fc2(self.fc1(x)))
```

```python
# 모델 인스턴스 생성
model = Net()
```

```python
# 모델 인스턴스 속성 확인
# 모델의 특정 층 추출

model.fc1.weight, model.fc2.bias
```

```python
# 모델의 각 층별 W, b 테넛 정보 확인
for i in model.parameters():  # 반환값이 Iterator(Generator)라면 반복문을 통하여 빼내야함
    print(i, end = '\n-------------------------------------------------------------------\n')
```

```python
# 각 층의 이름도 같이 확인
for i in model.named_parameters(): 
    print(i, end = '\n-------------------------------------------------------------------\n')
```

```python
# 모델 구성 모든 Layer 가져오기
for i in model.children(): 
    print(i,end = '\n\n')

for i in model.named_children():
    print(i, end='\n\n')
```

- Layer의 가중치 설정

```python
# 세비어 알고리즘의 가중치 초기화
nn.init.xavier_uniform_(model.fc1.weight)
```

```python
# 헤 알고리즘의 가중치 초기화
nn.init.kaiming_uniform_(model.fc2.weight)
```

```python
# torchinfo로 모델 구조 확인
from torchinfo import summary

summary(model, input_size=(1, 8))  # Param은 절편 갯수 포함
```

## [ 접근자 Accessor ]
- Series / DataFrame 의 데이터가 날짜시간(datetime), 문자열(object), 범주형(category),등의 경우 데이터 타입과 관련된 전용의 함수들 사용할 수 있도록 해주는 속성
- Series.str.문자열처리관련함수명()
- Seires.dt.날짜시간처리관련속성명
- Series.cat.범주형처리관련함수명()

(1) 모듈 로딩<hr>

```python
import pandas as pd
```

(2) 데이터 준비 <hr>

```python
datas = {"name":['홍 길동','이 나영','마 징가','베 토벤']   ## key값이 컬럼명
         ,'age':[10,21,73,89],
         'reg_date': ['2010/1/23','2020/4/23','2000/7/18','2017/12/3']}

datas2 = ["name",'age', 'reg_date'], ['홍 길동','이 나영','마 징가','베 토벤'],[10,21,73,89], ['2010/1/23','2020/4/23','2000/7/18','2017/12/3']
```

(3) 데이터 저장 <hr>

```python
# Dict ==> DataFrame
df1 = pd.DataFrame(datas)
df1
```

```python
# List ==> DataFrame
df2=pd.DataFrame(datas2)
df2
```

(4) 데이터 정보 확인 <hr>

```python
# 데이터의 전체 기본 정보 -> info()
df1.info()
```

(5) 데이터 전처리<hr>

- (5-1) 행 <-> 열 치환

```python
df1 = df1.T
df1 = df1.transpose()
```

(5-2) 실제 데이터와 타입 확인 후 형변환

```python
df1.head()
```

```python
# 1번 컬럼 : object => int
df1['age'] = df1['age'].astype('uint8')
```

```python
# 2번 컬럼 : object => datetime64[ns]
df1['reg_date'] = df1['reg_date'].astype('datetime64[ns]')
```

```python
df1.dtypes
```

(5-3) 0번 컬럼의 이름을 성과 이름으로 분리

```python
# 0번 컬럼만 추출
nameSR = df1['name']
nameSR, nameSR[1]   # nameSR[1] type : str
```

```python
# "이 나영" 데이터를 추출 => 성과 이름으로 분리
# nameSR[1].split()
```

```python
# for idx in range(nameSR.shape[0]):
#     nameSR[idx] = nameSR[idx].split()

# nameSR
```

```python
namesDF = nameSR.str.split(expand=True)
```

(5-4) 두개의 DataFrame을 컬럼 방향으로 연결

```python
df1 = pd.concat([df1,namesDF], axis = 1, ignore_index = True )  # 기본값은 axis = 0 -> 밑으로 연결됨
```

(5-5) 2번 컬럼에서 년,월,일 분리

```python
yearSR = df1[2].dt.year
```

```python
monthSR = df1[2].dt.month
```

```python
daySR = df1[2].dt.day
```

```python
pd.concat([df1,yearSR,monthSR, daySR], axis=1, ignore_index=True)
```

(5-6) 현재 DataFrame을 파일로 저장하기   =>  DF.to_확장자('파일경로+파일명)

```python
# CSV ==> 데이터와 데이터의 구분자
filename = './backup.csv'   # . : 현재위치   
#          ='backup.csv' 
df1.to_csv(filename, sep = ' ', index=False, header=False)  # index : 행 인덱스 포함 여부 # header : 열 인덱스 포함 여부  ==> 실제 데이터만 들어옴
```

# [ DataFrame 살펴보기 ]
* 데이터를 테이블 즉, 표 형태로 저장하는 자료구조
* 구성 : 행 + 열
* 생성 : pandas.DataFrame() 생성자 메서드

(1) 모듈로딩 ------------------------------------------------

```python
from pandas import DataFrame
```

(2) 데이터 준비 ------------------------------------------

```python
data = [ [11,22,33], [44,55,66],[77,88,99] ]
```

(3) DataFrame으로 저장 즉, DataFrame 객체 생성 ----------------

```python
df1 = DataFrame(data)
df1
```

(4) DataFrame의 속성 읽기 => 객체변수명.속성명 --------------------------------

```python
df1.index     # 행 갯수
```

```python
df1.columns     # 열 갯수
```

```python
df1.values
```

```python
# 데이터가 저장된 표의 형태
df1.shape       ## (행 수, 열 수)
```

```python
# 데이터의 차원 정보 n-dimention
df1.ndim
```

```python
# 컬럼별 데이터 타입 
df1.dtypes
```

(5) 원소 읽기 ==> 행과 열 지정 ==> 1개 원소 -------------------------------------------

- 컬럼 즉 열 (column) 읽기 ==> 객체변수명[인덱스]

```python
df1[0]    # df1[-1]은 안됨    ## type : serise
```

```python
df1[2]
```

```python
## 컬럼 변경 =========================================================
# 컬럼명 변경 ==> 객체변수명.columns = 새로운 컬럼명
df1.columns = ['A','B','C']     ## 원래 존재했던 위치 레이블 사용 불가
df1
```

```python
df1['A']                     ##df1.A 도 가능(문자일 때만 가능)
```

```python
# 컬럼 레이블 설정 후 정수 인덱싱 불가
# df1.0
```

- 로우 즉 행 (row) 읽기 ==> 방법 2가지
     * 위치 인덱스로 읽기 : 객체변수명.iloc[인덱스]
     * 레이블 인덱스로 읽기 : 객체변수명.loc[인덱스]


```python
df1.index = ['one','two','three']
```

```python
df1
```

```python
one = df1.loc['one']
type(one)             ## type : series
```

```python
df1.iloc[0]
```

(6) 모든 원소를 한번에 한개씩만 출력

```python
df1
```

```python
for r in df1.index:
    for c in df1.columns:
        print(df1.loc[r][c])
```

- 객체변수명.iloc[행 인덱스, 열 인덱스] / 객체변수명.loc[행 레이블, 열 인덱스 or 열 레이블]

```python
df1.iloc[0,0]
```

```python
df1.loc['one','A']             # df1.loc['one',0]도 가능 
```

```python
df1.loc['one','B'], df1.loc['two','C']
```

```python
df1.iloc[0,1], df1.iloc[1,2]      ## iloc는 정수형 인덱스만 가능
```

```python
df1.iloc[0][1]     # df1.iloc[0]은 serise -> serise의 1번 인덱스 값
```

```python
df1.loc['one']['B']
```

```python
zero = df1.iloc[0]
zero
```

```python
b = df1['B']
b
```

```python
df1.shape[0], df1.shape[1]       # 행 열
```

- iloc를 활용한 모든 원소 출력 ==> 행과 열의 번호

```python
for r in range(df1.shape[0]):
    for c in range(df1.shape[1]):
        print(df1.iloc[r,c])      # 정수형 인덱스를 반환하니까  iloc
```

- loc를 활용한 모든 원소 출력 ==> 행 레이블/이름과 열 레이블/이름

```python
for r in df1.index:
    for c in df1.columns:
        print(df1.loc[r,c])
```

<h1>시리즈(Seriese) 살펴보기</h1>  

 - 판다스에서 데이터를 저장하는 자료구조 중 하나  
 - 동일한 타입의 데이터를 연속된 메모리 공간에  저장 함  
 - 구성 : 인덱스 + 데이터  
 - 생성 : Series() 생성자 메서드  

### (1) 모듈 로딩 -------------------------------------

```python
import pandas as pd
```

### (2) 시리즈 객체 생성 ---------------------------------------

```python
# 객체 생성에 사용되는 메서드 => 생성자(Construtor)
# 데이터 타입과 동일한 이름의 메서드 => Series()
sr = pd.Series([11,22.,33])   ## shift + tab -> 함수 설명 보기 가능   ## 다른 타입들을 list에 넣으면 제일 메모리가 큰 타입을 기준으로 타입형변환
```

```python
sr
```

```python
type(sr)
```

### (3) 시리즈 객체의 속성(attributes)/필드(field) 살펴보기 ---------------------------------------
    * 기본 : data, index, dtype
    * 읽기 : 객체변수명.속성명
    * 쓰기 : 객체변수명.속성명 = 새로운 값

```python
# 현재 시리즈 객체에 저장된 인덱스 속성 확인
sr.index
```

```python
# 현재 시리즈 객체에 저장된 데이터 즉 값 확인
sr.values
```

```python
# 시리즈 객체의 인덱스 변경
sr.index = ['A','B','C']    ## 전체 갯수와 똑같아야함
```

```python
sr
```

```python
# values 속성은 속성방식으로 변경 불가!!
#sr.values = [777,888,999]
```

## [다양한 Series 객체 생성]

 #### (1) 모듈 로딩

```python
import pandas as pd
```

 #### (2) Series 객체 생성 - 방법 1
  - Series( data, index = 지정)

```python
sr1 = pd.Series([11,22,33], index = ['one','two','three'])   ##인덱스 지정하고 객체 생성
```

```python
sr1
```

 #### (2) Series 객체 생성 - 방법 1

```python
sr2 = pd.Series([11,22,33], index = ['one','two','three'], name = 'jumsu')
```

```python
sr2
```

 #### (2) Series 객체 생성 - 방법 3

```python
sr1 = pd.Series([11,22,33], index = ['one','two','three'], name = 'jumsu', dtype ='uint8')
sr1
```

 #### (2) Series 객체 생성 - 방법 4 => 저장되는 데이터 dict

```python
data = {'name':'홍길동','age':12,'job':'도적'}   ## key가 인덱스 이름으로 생성됨
sr1=pd.Series(data)
sr1
```

```python
# 속성 확인하기
print(sr1.index, sr1.values, sep = '\n')   ## 데이터 타입이 섞이면 데이터 손실 없게 하기 위해 타입을 object로 맞춰버
```

 #### (2) Series 객체 생성 - 방법 5 => 저장되는 데이터 set

```python
data = {'홍길동',12,'도적'}    # 순서 없는 데이터, 중복 불허

# 형변환 후 객체 생성 가
sr1 = pd.Series(list(data))
sr1
```

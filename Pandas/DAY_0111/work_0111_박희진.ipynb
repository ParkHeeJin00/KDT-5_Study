(6) 컬럼별 데이터들의 속성 즉, index, values, shape, ndim, type 정보 출력

```python
for colname in movieDF.columns:
    print(movieDF[colname].index)
    print(movieDF[colname].values)
    print(movieDF[colname].shape)
    print(movieDF[colname].ndim)
    print(movieDF[colname].dtype)
```

(7) 컬럼별 인덱스를 변경 실습

```python
movieDF.index = [f'{i}번째' for i in movieDF.index]
```

(8) 컬럼별 데이터 즉 값 변경 실습

(9) DataFrame의 원소를 모두 출력 단, 2가지 방법 사용

```python
# 1
for r in movieDF.index:
    for c in movieDF.columns:
        print(movieDF.loc[r,c])
```

```python
# 2
for r in range(movieDF.shape[0]):
    for c in range(movieDF.shape[1]):
        print(movieDF.iloc[r,c]) 
```

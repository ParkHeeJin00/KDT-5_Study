### [ Series 객체 원소 인덱싱 & 슬라이싱]

(1) 모듈 로딩 ----------------------------------------------------------

```python
from pandas import Series
```

(2) 데이터 준비 ------------------------------------

```python
datas = {'name':'홍길동','age':12,'job':'도적'}
```

(3) Series 객체 생성 ------------------------------------------

```python
sr1 = Series(datas)
```

```python
sr1
```

(4) 인덱싱으로 원소 읽기 --------------------------------

```python
sr1['age'],sr1[1],sr1.index
```

=> 2개 원소 읽기

```python
# 이름과 직업 데이터 읽기 ==> 원소 단위로 읽기
sr1['name'], sr1['job'],sr1[0],sr1[2],type(sr1[0])   ## type - str   ## 슬라이싱하면 type이 변함
```

```python
# 이름과 직업 데이터 읽기 ==> 리스트 단위로 읽기
ret1 = sr1[['name','job']]   ##레이블인덱스
ret2 = sr1[[0,2]]            ##위치인덱스

type(ret1), ret1       ## type - Seires
```

```python
type(ret2), ret2
```

(5) 슬라이싱으로 원소 읽기 --------------------------------

- 위치 인덱스로 슬라이싱 -------------------------------

```python
# 위치 인덱스로 슬라이싱 -------------------------------
sr1[:2]         # 시작<= ~ <끝
```

```python
sr1[:]          # 시작<= ~ <=끝
```

```python
sr1[:-1]        ##음수 인덱스 가능
```

- 레이블 인덱스로 슬라이싱 ------------------------------

```python
sr1['name':'job']          # 시작<= ~ <=끝
```

```python
sr1['name':'age']
```

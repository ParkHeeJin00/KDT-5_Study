### [ 데이터 필터링 ]
- 원하는 데이터를 추출하기 위해서 사용하는 방법
- 원하는 데이터 걸러내겠다는 의미
- 다양한 조건 검사 진행 => 비교 검사 : 같다 ==, 같지 않다 !=, 크다 >, 크거나 같다 >=, 작다 <, 작거나 같다 <=

(1) 모듈 로딩 <hr>

```python
import pandas as pd
import numpy as np
```

(2) 데이터 준비 <hr>

```python
file = "../../DATA/employees.csv"
```

(3) 데이터 저장 <hr>

```python
empDF = pd.read_csv(file)
```

(4) 데이터 확인 <hr>

(5) 데이터 필터링

- [문제] employees 데이터에서 이름이 Maria인 데이터만 추출

- bool값으로 구성된 데이터 ==> 불린 인덱싱(Boolean Indexing)

```python
mariaMask = empDF['First Name'] == 'Maria'   
## 브로드 캐스팅 : empDF와 사이즈가 같은데 모든 값이 Maria인 series 생성
type(mariaMask),mariaMask.sum()     ## series type
```

```python
empDF[empDF['First Name'] == 'Maria']   ## True에 해당하는 행만 반환
## = empDF.loc[empDF['First Name'] == 'Maria']   =   empDF[mariaMask]
```

- [문제] employees 데이터에서 연봉이 50000 이상이 데이터 추출

```python
empDF['Salary']>= 50000   ## 인덱스로 사용
```

```python
empDF[empDF['Salary']>= 50000]  # True인 행만 추출
```

- [문제] employees 데이터에서 성별이 남자이며 연봉이 100000이상인 데이터 조회 출력

```python
GenderMask = empDF['Gender'] == 'Male'
```

```python
SalaryMask = empDF['Salary'] >= 148000
```

```python
empDF[GenderMask & SalaryMask]
```

- 여러 개의 조건이 모두 True여야만 하는 조건 ==> AND 조건 (&)
    * 문 법 : 조건1 & 조건2 & 조건3 &...& 조건N
    * 연산자 : & 앰퍼센트
    * 결 과 : True / False

```python
mask = ( (empDF['Gender'] == 'Male') & (empDF['Salary'] >= 148000) )
type(mask), mask.sum()  ## type : series
```

```python
empDF[mask]
```

- 여러 개의 조건 중 하나가 True면 True가 되는 조건 ==> OR 조건 (|)
    * 문 법 : 조건1 | 조건2 | 조건3 | ...| 조건N
    * 연산자 : | 파이프
    * 결과 : Ture / False

- [문제] employees 데이터에서 입사일이 1993년 1월 1일 이전이거나 또는 팀이 없는 사람 데이터 조회

```python
mask = (empDF['Mgmt'] == False) | (empDF['Start Date'] < '1993-01-01')
```

```python
pd.to_datetime(empDF['Start Date'])    ## datetime으로 변환
```

```python
empDF[mask]
```

```python
(empDF['Team'] == 'Sales')|(empDF['Team'] == 'Legal')|(empDF['Team'] == 'Marketing')
```

```python
empDF['Team'].isin(['Sales','Legal','Marketing']) 
```

- 구간 지정 검사 불린 메서드 => DF.between(low,high)

```python
(empDF['Salary'] >= 80000) & (empDF['Salary'] < 90000)
```

```python
empDF['Salary'].between(80000,90000)
```

```python
empDF['First Name'].str.upper()    ## str : 지시자 / 해당 타입일때 함수와 연결해주는 것
```

## [ 중복 데이터 처리 ]
- 중복 데이터 기준 ==> 행(row)
- 행 단위로 모든 컬럼의 값이 완전히 똑같은 데이터는 중복 데이터
- 처리 => 제거

(1) 모듈 로딩 <hr>

```python
import pandas as pd
```

(2) 데이터 준비 <hr>

```python
file = '../../DATA/employees.csv'
```

(3) 데이터 저장 : CSV => DF<hr>

```python
empDF = pd.read_csv(file)
```

(4) 데이터 확인 <hr>

```python
empDF.info()
```

(5) 데이터 전처리 => 나. 중복 데이터 처리<hr>

- 중복 데이터 체크 => duplicated() : 중복이면 True
- 중복 데이터 처리 => drop_duplicates() : 삭제

```python
# 행 단위로 완전히 똑같은 값이 있는지 검사 ==> 0개
empDF.duplicated().sum()
```

```python
# Team 컬럼
empDF.Team.duplicated()   ## 이전 값에서 똑같은 값이 있었다면 True
# = empDF.duplicated(subset=['Team'])

empDF.Team.duplicated().sum(), empDF.Team.nunique()
```

```python
empDF['Team'].drop_duplicates(keep=False)  
## keep = 'first'(기본값 : 처음거 놔둠) or 'last'(마지막거 놔둠) or False(모두 삭제) => 어디서 부터 확인하고 살려둘지
# 바꾸는건 앵간하면 복사본 사용해라~~
```

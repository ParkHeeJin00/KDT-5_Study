- 결측치 처리 ==> (1) 삭제 : dropna()

```python
## [기본] 모든 행에 1개의 값이라도 NaN이면 삭제

empDF2 = empDF.dropna()     ## 반환값 : 복사본
print(empDF.shape, empDF2.shape)
```

```python
## [설정] 모든 행의 값이 NaN이면 삭제

empDF3 = empDF.dropna(how = 'all')
print(empDF.shape, empDF3.shape)
```

```python
## [설정] 특정 컬럼의 Na만 체크 후 삭제하는 방법

# 예) 성별에 따른 연봉 분석 => NA면 분석 불가능 컬럼을 지정
empDF4 = empDF.dropna(subset=['Gender','Salary'])   # subset = 리스트
print(empDF.shape, empDF4.shape)
```

```python
## [설정] 컬럼별 정상 데이터 즉 NA가 아닌 데이터의 갯수(컬럼 별 데이터 갯수)를 지정 후
#         해당 수 만큼 정상 데이터가 없는 컬럼 삭제

empDF5 = empDF.dropna(thresh=900, axis=1)   ## Gender의 정상데이터 갯수는 854개 => 900미만 삭제 => Gender 컬럼 삭제  
print(empDF.shape, empDF5.shape)

# 적어도 20% 미만의 결측치를 가진 데이터만 사용하고 싶다면 전체 데이터 갯수가 1000일 때, thresh = 800 으로 지정
```

- 결측치 처리 => 대체 :fillna()

```python
# Gender 컬럼의 결측 값을 무엇으로 채울지??
empDF.describe(include='all')
```

```python
# ==> 컬럼의 최빈값으로 na 채우기
empDF.Gender.mode()[0]    # empDF.Gender.mode()의 type : Series  =>  값을 지정할 때 인덱스 이용
genSR = empDF.Gender.fillna(empDF.Gender.mode()[0])
genSR.isna().sum()   # => 0
```

```python
# ==> NA값 이전과 이후 값으로 채우기 => method = 'ffill' or method = 'bfill'
empDF.Gender.fillna(method='bfill')   ## bfill로 채우면 마지막값 NaN값
```

```python
empDF.Gender.mode()
```

```python
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                    [3, 4, np.nan, 1],
                    [np.nan, np.nan, np.nan, np.nan],
                    [np.nan, 3, np.nan, 4]],
                    columns=list("ABCD"))   # 하나씩 뜯어서 들어감
```

```python
df
```

```python
df.A.fillna(method = 'ffill', limit = 1)   # 이전거를 채우되 한번만 채우겠다 -> 채우는 최대 갯수
```

```python
df.A.fillna(method = 'bfill')  # 뒤 값이 NaN값이라서 2, 3 값 안 채워짐 
```

```python
# 컬럼별로 원하는 값으로 채울 수도 있음

values = {"A": 0, "B": 1, "C": 2, "D": 3}
df.fillna(value=values)
```

```python
# DF으로 채울 수도 있음

df2 = pd.DataFrame(np.zeros([4, 4]), columns=list("ABCE"))  # zeros((4,4)) => 4행 4열로 만들어진 DF에 값을 모두 0으로 채워라
## But, 컬럼 값 달라서 채워줄 수가 없음 => df의 D 컬럼을 변화 X
df.fillna(df2)
```

```python
# B 컬럼과 D 컬럼을 평균으로 결측치를 대체   / 이상치에 대한 대비 X => 중앙값으로 대체하는 것이 나을수도 O

df.B.fillna(df.B.mean())
df.D.fillna(df.D.mean())
```

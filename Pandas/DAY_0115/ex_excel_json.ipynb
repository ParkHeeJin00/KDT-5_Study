## [ JSON => DataFrame 저장하기]

(1) 데이터 준비

```python
file3 = '../../DATA/nobel.json'
```

(2) JSON ==> DataFrame으로 저장

```python
nobelDF = pd.read_json(file3)
nobelDF.info()
```

(3) 데이터 확인

```python
print(nobelDF.head(3),nobelDF.tail(3), sep = '\n\n')
```

(4) 데이터 추출

```python
# 행 추출 => iloc[] / loc[]
nobelDF.iloc[0]     # type : series          / prizes : 행 레이블
```

```python
nobelDF.iloc[0]['prizes']    # type : dict
```

```python
# 행 추출 => iloc[n1,....] / loc[]
nobelDF.iloc[[0,3]]     # type : DataFrame
```

```python
oneData = nobelDF.iloc[0]['prizes']
type(oneData)
```

```python
oneData.keys(), oneData.values(), oneData.items()
```

```python
for k, v in oneData.items():
    print(f'{k} :\n=====>{v}')
    if k == 'laureates':  # key가 laureates인 value는 리스트라서 한번 더 풀기
        for d in v:  # v : dict 원소로 구성된 리스트   d : 리스트 내에 요소로 존재하는 dict 타입 데이터
            for k1,v1 in d.items():  # d : dict
                print(k1,v1)
```

- json 중첩 데이터의 다중 반복문 ==> 단일 리스트 형태 처리 : 평탄화 / 정규화

```python
pd.json_normalize(oneData)
```

```python
pd.json_normalize(oneData, record_path='laureates')  ##record_path : key가 컬럼명으로
## = pd.json_normalize(oneData['laureates'])
```

```python
## 여기서 컬럼 추가
pd.json_normalize(oneData, record_path='laureates', meta = ['year','category'])
```

- 모든 행에 대한 평탄화/정규화 진행

```python
# 컬럼 읽기 => DF객체변수명['컬럼명'], DF객체변수명.컬럼명
# pd.json_normalize(nobelDF.prizes, record_path='laureates')  => Error
# =====> 모든 행에 record_path에 해당하는 키(key) 존재해야 함!!
# =====> 없는 행에 record_path에 해당하는 키 추가

for data in nobelDF.prizes:
    print(type(data), data.keys(), end = '\n\n')
    # 해당 dict에 laureates 키가 없다면 추가
    data.setdefault('laureates', [])   # 기본값을 가지고 키 추가 : 있으면 냅두고 없으면 [] 넣어라
```

```python
nobelDF2 = pd.json_normalize(nobelDF.prizes, record_path='laureates', meta=['year','category'])  ## 반환값 DF => 저장해야함
```

- JSON의 다양한 방식의 DataFrame 저장하기 => orient 파라미터

```python
json_data='''
[
{"id" : "1","name":"sravan","age":22},
{"id" : "2","name":"harsha","age":22},
{"id" : "3","name":"deepika","age":21},
{"id" : "4","name":"jyothika","age":23}
]
'''   ## 파이썬에서는 list지만 제이슨에서는 array -> 한 줄을 한개로 봄
```

- orient = 'record' 설정 => JSON 파일 데이터를 행 단위로 읽어서 DataFrame에 저장

```python
df = pd.read_json(json_data, orient='record')   ## 컬럼 -> 밸류로
df
```

```python
# 데이터 => 키 : {값}
json_data2 = '''
{
    "student-1" : {"id" : "1","name":"sravan","age":22},   # 한 줄이 행 = 레코드 / 총 네개의 레코드 총 네개의 제이슨 객체
    "student-2" : {"id" : "2","name":"harsha","age":22},
    "student-3" : {"id" : "3","name":"deepika","age":21},
    "student-4" : {"id" : "4","name":"jyothika","age":23}

}
'''   ## 파이썬에서는 dict 제이슨에서는 dict like
## 키 -> 인덱스 / 값의 키 -> 컬럼 / 값의 값 -> 밸류
```

```python
# 키 => 인덱스로 설정
df2 = pd.read_json(json_data2, orient='index') # dict의 키를 인덱스로 설정
df2
## == pd.read_json(json_data2).T
```

```python
# 키 => 컬럼으로 설정
df3 = pd.read_json(json_data2, orient='columns')
df3
```

```python
# JSON Format 
json_data4 = '''
[
    [ "1", "sravan",22 ],   # 한 줄이 행(레코드)으로 들어옴
    [ "2", "harsha",22 ],
    [ "3", "deepika",21 ],
    [ "4", "jyothika",23 ]
]
'''   ## 모두 값으로 읽어옴
```

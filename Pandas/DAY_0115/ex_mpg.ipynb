## [ 데이터 파일 읽고 분석 ]

- (1) 모듈 로딩

```python
import pandas as pd
```

- (2) 데이터 준비

```python
file = '../../DATA/auto-mpg.csv'
```

- (3) 데이터 저장

```python
# 정규표현식 => \s+ : white space  즉, 공백 문자 1개 이상으로 된 문자가 구분자
mpgDF = pd.read_csv(file, header=None, sep='\s+')   
mpgDF
```

- (4) 데이터 확인

```python
# 전체 데이터에 대한 요약 정보
mpgDF.info()
```

```python
# 실제 데이터 확인 ===> 3번 컬럼이 실제 데이터 타입과 다름!
mpgDF.head()
```

- 컬럼별 데이터 수 확인 => DF.count()

```python
mpgDF.count()
```

- 컬럼별 고유한 데이터 갯수 확인 =>  DF.value_counts() 

```python
mpgDF.value_counts()   ## Series 리턴
```

```python
for col in mpgDF.columns:
    print(col, mpgDF[col].value_counts(),end = '\n\n')
```

```python
mpgDF[3].value_counts(dropna=True)  # 기본값
```

- 컬럼별 고유값 확인

```python
for col in mpgDF.columns:
    print(col,mpgDF[col].unique())   ## 고유한 값 출력
```

- 수치 데이터의 통계값 확인 => describe()

```python
mpgDF.describe()   # 표준편차가 작을수록 조밀
```

- 최빈값 계산 메서드 => mode()

```python
mpgDF.mode(numeric_only=True)  ## 컬럼4는 최빈값이 총 4로 동률값
```

- 중앙값 계산 메서드 => median()

```python
mpgDF.median(numeric_only=True)
```

- 컬럼과 컬럼 사이의 관계의 수치화 해주는 메서드 => corr()

```python
mpgDF.corr(numeric_only=True) # 0번 컬럼은 4번 컬럼과 음의 상관관계를 갖는다.
```

- 컬럼명 지정

```python
mpgDF.columns = ['mpg','cylinders','displacement','horsepower','weight','aceleration','model year','origin','car name']
```

- 데이터 타입과 실제 타입이 다른 컬럼 ==> 타입 캐스팅

```python
# '?'을 어떻게 처리 할건지
mpgDF['horsepower'].value_counts()['?']  # '?' 개수 확인
# mpgDF['horsepower'].value_counts() -> Series 객체
```

```python
horseSR = mpgDF['horsepower'].copy()
```

```python
for idx, horse in enumerate(horseSR):   ## series에서 원소 읽으려면 인덱스 필요 => enumerate 함수 사용
    if horse =='?':
        horseSR[idx] = '0.0'
```

```python
mpgDF['horsepower'] = horseSR.astype('float64')
```

- 모든 컬럼과 mpg 연비의 관계 비교

```python
mpgDF.corr(numeric_only=True).mpg
```

```python
mpgDF.corr(numeric_only=True).mpg.sort_values(ascending=False,key= lambda v:v.abs())
```

```python
mpgDF.info()
```

- origin 컬럼의 타입을 int64 => category로 타입 캐스팅

```python
mpgDF['origin'] = mpgDF['origin'].astype('category')
```

```python
mpgDF.info()
```

```python
## [ 데이터 원소값 변경/ 타입 변경 동시에 진행 ]
```

```python
mpgDF2 = pd.read_csv(file, header=None, sep='\s+', names=['mpg','cylinders','displacement','horsepower','weight','aceleration','model year','origin','car name'])   
```

```python
# 데이터 정보 확인
mpgDF2.info()
```

```python
# 원소값 '?' 변경해주는 메서드 => replace(old,new)
mpgDF2.horsepower.replace('?','0.0', inplace=True)    # 기본값은 copy본 반환
```

```python
# origin 컬럼의 1 - USA, 2 - EU, 3 - JAPAN
# mpgDF2.origin.replace(1,'USA',inplace=True)
# mpgDF2.origin.replace(2,'EU',inplace=True)
# mpgDF2.origin.replace(3,'JAPAN',inplace=True)

# dict형도 가능
# mpgDF2.origin.replace({1:'USA',2:'EU',3:'JAPAN'}, inplace=True)

# list형도 가능
mpgDF2.origin.replace([1,2,3],['USA','EU','JAPAN'], inplace=True)
```

```python
# origin 컬럼의 값의 종류별 갯수 확인 => value_counts() / unique()
mpgDF2.origin.value_counts()
```

```python
mpgDF2.origin.unique()
```

```python
mpgDF2.origin.nunique()
```

- horsepower 컬럼과 origin 컬럼의 데이터 타입 변경

```python
mpgDF2.horsepower = mpgDF2.horsepower.astype('float64')
```

```python
mpgDF2.origin = mpgDF2.origin.astype('category')
```

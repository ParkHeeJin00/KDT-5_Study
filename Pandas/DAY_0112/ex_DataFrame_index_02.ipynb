## [ DataFrame 인덱스 다루기 ]

- (1) 모듈 로딩 <hr>

```python
import pandas as pd
```

- (2) 데이터 준비 <hr>

```python
# dict 타입의 데이터 생성
exam = {'이름':['박희진','손예림','명노아','전진우']   
        ,'국어':[100,80,40,39]   
        ,'미술':[10,30,40,39]
        ,'수학':[92,88,79,58]
        ,'체육':[92,68,99,88]}
```

- (3) 데이터 저장<hr>

```python
examDF = pd.DataFrame(exam)
```

- (4) 데이터 확인<hr>

```python
examDF
```

- 행 데이터 추가

```python
# 국어 점수 같게
examDF.loc[4] = ['박희진',100,90,92,58]
```

```python
# 체육 점수가
examDF.loc[5] = ['박희진',100,70,60,90]
```

- (5) 새로운 인덱스 설정 - 방법 3 : 멀티인덱스 제어 DF.set_index([컬럼1,컬럼2,...,컬럼n])<hr>

```python
# 이름, 국어, 수학을 인덱스 설정
examDF1 = examDF.set_index(['이름','국어','수학'])
```

```python
examDF1.index
```

- 행 데이터 선택 : DF객체변수명.loc[행 레이블] / DF객체변수명.iloc[위치 인덱스]

```python
examDF1.loc['박희진',100,92]
```

```python
examDF1.loc['박희진',100,60]
```

```python
# Key Error 발생 ==> ('박희진',100,92)
# exam.loc[(100)]
```

- 멀티 인덱스의 일부분을 기준으로 행 선택 : xs
    * 예) '박희진' 100 92 => Level 0 1 2

```python
examDF1.xs('박희진')        ## = examDF1.xs('박희진', level = 0)  
```

```python
examDF1.xs(100, level = 1)   
```

```python
examDF1.xs(92, level = 2) 
```

- 2개 행을 한꺼번에 가져오기

```python
examDF1.loc[[('박희진',100,92),('명노아',40,79)]]
```

- (5) 인덱스 재배치 : 행/열 인덱스를 원하는 순서대로 재배치 + 존재하지 않는 행/열 추가
    * DF.reindex()

```python
print(examDF.index, examDF.columns, sep = '\n\n')
```

- 컬럼명 순서 변경 ==> '체육','미술','국어','수학'

```python
examDF.reindex(['체육','미술','국어','수학','총합'], axis='columns', fill_value=0)   ## 안에 안 적은 col은 사라짐 + 적은 col 생성됨
## fill_value : 새로 생긴 col에 넣을 값을 지정
```

```python
examDF.reindex([0,2,4,1,3,5], axis='index',  fill_value=0)
```

- 컬럼명 순서 변경 ==> '체육','미술','국어','수학' + 행 인덱스 변경 => 0,2,4,1,3,5 재배치

```python
examDF.reindex(index = [0,2,4,1,3,5], columns=['체육','미술','국어','수학','총합'], fill_value=0)
```

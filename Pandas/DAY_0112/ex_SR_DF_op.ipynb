- 논리연산 => & : and 연산 / | : or 연산 / ~ : not 연산

```python
## and , or, not 판다스에서는 안됨
(sr > 5) & (sr < 10)
```

```python
~(sr > 5 )   ## 5보다 작거나 같으면 True 반환
```

- (5) 연산 수행 => Series 객체 + Series 객체

```python
data2 = [2,4,6,8,9]

sr2 = pd.Series(data2)
```

```python
sr2
```

```python
print(sr+sr2, sr-sr2, sr* sr2, sr%sr2, sep = '\n\n')
```

```python
print(sr > sr2, sr < sr2, sep = '\n\n')
```

```python
sr2.index = [2,3,4,5,6]
```

```python
print(sr+sr2, sr-sr2, sr* sr2, sr%sr2, sep = '\n\n')    ## index가 같은 것끼리 연산
```

- (5) Series 연산수행 => 전용 메서드

```python
sr.add(5, fill_value=0)     # Nan 값에서 0 넣고 연산
```

```python
sr.sub(5, fill_value=0)     # Nan 값에서 0 넣고 연산
```

```python
sr.mul(sr2, fill_value=0)
```

```python
addSR = sr.add(sr2, fill_value=0)
subSR = sr.sub(sr2, fill_value=0)
mulSR = sr.mul(sr2, fill_value=0)
divSR = sr.div(sr2, fill_value=0)
```

```python
addSR, subSR, mulSR, divSR
```

- (6) DataFrame 연산 수행

```python
df = pd.DataFrame([addSR,subSR,mulSR,divSR], index = ['add','sub','mul','div'])
df
```

- DataFrame과 숫자 연산 수행

```python
print(df+5, df-5, sep = '\n\n')    ## 5로 다 채워진 Serise 생성 : 브로드캐스팅
```

```python
print(df>5,df<5, sep = '\n\n')
```

- DataFrame과 DataFrame 연산 수행 => 동일 위치의 원소끼리 연산 수행

```python
df2 = df*10
df2
```

```python
print(df+df2, df-df2, df*df2, df%df2, sep = '\n\n')    ## DataFrame 4개 생성    # col명과 row명이 같아야 위치 인식하고 계산
```

```python
df2.index = ['add','minus','mul','/']
# col명과 row명이 같아야 위치 인식하고 계산 / 인덱스 명 바꾸니까 계산 안됨 , 확장해서 연산 수행 => 연산 결과 : NaN 값
```

- 연산 수행 메서드 활용 => 결측치 즉 NaN 빈칸에 대한 처리를 수행 가능   &
    * fill_value 파라미터에 값 설정

```python
df.add(df2, fill_value=0)
```

- (5) 데이터 정렬
    * 인덱스 기반 정렬 => 행 인덱스 / 열 인덱스

```python
# 행 인덱스 기반 정렬 => DF.sort_index()
movieDF.sort_index()    # ==> 오름차순 정렬     ## 내림차순 정렬 / 오름차순 정렬(기본값/True)
```

```python
# 컬럼 인덱스 기반 정렬 => 내림차순
movieDF.sort_index(axis='columns',ascending=False)    ## 열 : axis = 1   / 행 : axis = 0
```

- (5) 데이터 정렬

```python
movieDF.sort_values(by = ['Year','Title']) 
## Year값을 기준으로 정렬하고 같은 정렬 값을 가질 때, Title 값으로 정렬
```

- [ 실습 ] 다양한 정렬


```python
import numpy as np                     # nan : Not a Number 빈칸 의미
from math import nan
data={'col1':['A','A','B',nan,'D','C'],
      'col2':[2,1,9,8,7,4],
      'col3':[0,1,9,4,2,3],
      'col4':['a','B','c','D','e','F'],}

dataDF=pd.DataFrame(data)
dataDF
```

- col 컬럼을 기준으로 오름차순 정렬

```python
# 정렬할 때, NoN 즉, 빈칸 데이터 (결측치) 값의 위치 설정 => na_postion (기본값 : 'last' , 제일 마지막)
dataDF.sort_values(by='col1',na_position='first')
```

```python
# col4 컬럼을 기준으로 오름차순 정렬
dataDF.sort_values(by = 'col4')
```

```python
# col4 컬럼을 기준으로 오름차순 정렬 / 대 소문자 구별 없이 
dataDF.sort_values(by = 'col4', key=lambda col:col.str.upper())   # col은 object
                                # 데이터를 함수 사용해 변경하고 정렬하고 싶을 때, key 파라미터 사용
```

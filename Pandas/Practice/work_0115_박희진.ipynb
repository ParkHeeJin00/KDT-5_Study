2. 시도별 전출입
- 전출 -> 전입 인구 분석
- 서울과 제주로 가장 많이 이주한 지역

```python
moveDF = pd.read_excel('../DATA/시도별 전출입 인구수.xlsx', index_col=[0,1], header = [0,1])
```

```python
moveDF
```

```python
moveDF.replace('-','0', inplace=True)
```

```python
moveDF.info()
```

```python
for i in moveDF.columns:
    moveDF[i] = moveDF[i].astype('int64')
```

```python
moveDF.info()
```

```python
# 전출 -> 전입 인구 분석
moveDF.corr()
```

```python
moveSDF = moveDF.xs('서울특별시', level = 1)
moveJDF = moveDF.xs('제주특별자치도', level = 1)
```

```python
sumS = {}
for i in moveSDF.index:
    if i != '전국' and i != '서울특별시':    ## update가 안되는 이유??
        sumS[i] = moveSDF.loc[i].sum()
sumS
```

```python
for k, v in sumS.items():
    if v == max(sumS.values()):
        print(f'{k} 지역이 서울로 가장 많이 이주한 지역입니다.')
```

```python
sumJ = {}
for i in moveJDF.index:
    if i != '전국' and i != '제주특별자치도':
        sumJ[i] = moveJDF.loc[i].sum()
sumJ
```

```python
for k, v in sumJ.items():
    if v == max(sumJ.values()):
        print(f'{k} 지역이 제주도로 가장 많이 이주한 지역입니다.')
```

```python
## 함수 구현

def moveCityMax(city):

    movecDF = moveDF.xs(f'{city}', level = 1)
    
    sumcDF = {}
    for i in movecDF.index:
        if i != '전국' and i != city:   
            sumcDF[i] = movecDF.loc[i].sum()

    for k, v in sumcDF.items():
        if v == max(sumcDF.values()):
            print(f'{k} 지역이 {city}로 가장 많이 이주한 지역입니다.')

moveCityMax('경상북도')
```

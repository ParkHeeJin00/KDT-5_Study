```python
file = pd.read_csv('../DATA/revolutionary_war.csv')
file  # DataPrame
```

```python
sr = pd.Series(list(file['Start Date']))
sr = pd.to_datetime(sr)
```

```python
type(sr)
```

```python
def day(date):
    return date.strftime('%A')
```

```python
sr.dropna().apply(day)  # 결측치 제거한 값들에 동시에 함수 적용(dropna -> apply)
```

```python
sr.dropna().apply(day).value_counts()
```

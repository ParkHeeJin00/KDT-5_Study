[문4] 아래 조건을 만족하는 데이터 객체를 생성하세요.
1) 데이터 : stock price.xlsx
2) 데이터 로딩 시 인덱스로 사용할 수 있는 컬럼 설정
3) 결측치 처리 =>
   (가) 1개라도 NaN이면 모두 제거 / price가 NaN이면 제거 / value가 10000이하면 제거
   (나) 평균값으로 결측치 대체, 이전/이후 값으로 결측치 대체
4) value가 1000이하 이거나 price가 90000이상인 데이터

```python
stockDF = pd.read_excel('../DATA/stock price.xlsx', index_col=[0])
stockDF
```

```python
# (가)
stockDF2 = stockDF.dropna()
stockDF2 = stockDF.dropna(subset='price')
stockDF2 = stockDF[stockDF['value']<=1000].dropna()
```

```python
# (나)
stockDF.isna().sum()  # value에 두개 price에 두개
# 평균으로 대체
stockDF3 = stockDF['value'].fillna(stockDF['value'].mean())
stockDF3 = stockDF['price'].fillna(stockDF['price'].mean())

# 이전값으로 대체
stockDF4 = stockDF['value'].fillna(method='ffill')
stockDF4 = stockDF['price'].fillna(method='ffill')

# 이후값으로 대체
stockDF5 = stockDF['value'].fillna(method='bfill')
stockDF5 = stockDF['price'].fillna(method='bfill')
```

```python
stockDF3.isna().sum()
stockDF4.isna().sum()
stockDF5.isna().sum()
# 결측치 없는거 확인 가능
```

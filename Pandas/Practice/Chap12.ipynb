```python
import pandas as pd
```

```python
tv_shows_json = pd.read_json('../DATA/tv_shows.json')
tv_shows_json  ## shows 중첩 -> episode 중첩
```

```python
tv_shows_json.iloc[0][0]  #확인
```

```python
tv_show = pd.json_normalize(data = tv_shows_json['shows'], record_path='episodes', meta=['show','runtime','network'])
```

```python
tv_show
```

- 2번

```python
xtv = tv_show[tv_show["show"] == 'The X-Files']
xtv
```

```python
losttv = tv_show[tv_show["show"] == 'Lost']
losttv
```

```python
buftv = tv_show[tv_show["show"] == 'Buffy the Vampire Slayer']
buftv
```

- 3번

```python
epidata = pd.ExcelWriter("episodes.xlsx")
```

```python
xtv.to_excel(excel_writer=epidata, sheet_name="The X-Files", index=False)
```

```python
losttv.to_excel(excel_writer=epidata, sheet_name='Lost', index=False)
```

```python
buftv.to_excel(excel_writer=epidata, sheet_name='Buffy the Vampire Slayer', index=False)
```

```python
epidata.save()
```

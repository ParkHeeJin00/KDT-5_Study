### 이미지 데이터 분류
- 데이터 : sklearn.datasets의 MNIST ==> 디지털 숫자 0 ~ 9 손글씨 데이터셋
- 피쳐 : 이미지 픽셀 수, 28 * 28 = 784
- 타겟 : 0 ~ 9, 총 10개
- 학습유형 : 지도학습 + 분류

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
# 모듈 로딩
from sklearn.datasets import fetch_openml
```

```python
# 데이터 로딩
mnist = fetch_openml('mnist_784',parser='auto') # 번치 타입
```

```python
mnist.data.shape , mnist.target.shape
```

```python
mnist.data.describe()
```

[2] 데이터 전처리 <hr>
- 이미지 픽셀값 0~1로 스케일링
- MinMax써도 됨

```python
featureDF = mnist.data
featureDF = featureDF / 255 # 최대값으로 나눈 것
featureDF.describe()
```

```python
targetSR = mnist.target
```

```python
targetSR.unique()  # 문자열  
```

```python
targetSR = targetSR.astype('int')
```

[3] 기계학습용 데이터셋 준비 <hr>
- 학습용, 검증용, 테스트용 3개 데이터셋 준비

```python
# 모듈 로딩
from sklearn.model_selection import train_test_split
```

```python
# 학습용 & 테스트용 분리 => MNIST database에서 제공한 6만개와 만개
xtrain,xtest,ytrain,ytest = featureDF.iloc[:60000], featureDF.iloc[60000:], targetSR[:60000], targetSR[60000:]

xtrain.shape,xtest.shape
```

```python
# 학습용 & 검증용 분리 => 학습용 데이터셋에서 일부를 검증용으로 준비
# 검증용 : 학습 시 모델 최적화가 잘 되는지 검사하는 데이터셋
# 테스트용 데이터 스코어가 높아야함. 검증용엔 중간에 평가할 때 사용
# 검증용 데이터는 매번 바뀌는게 좋음
xtrain,xval,ytrain,yval = train_test_split(xtrain,ytrain,
                                           test_size=0.3,
                                           stratify=ytrain,
                                           random_state=5)
xtrain.shape,xval.shape
```

[4] 학습진행 <hr>

[4-1] 분류 방법 : : LogisticRegression

```python
from sklearn.linear_model import LogisticRegression, SGDClassifier
```

```python
lr_model = LogisticRegression(max_iter=3000)
lr_model.fit(xtrain,ytrain)
```

```python
train_score = lr_model.score(xtrain,ytrain)
val_score = lr_model.score(xval,yval)

train_score, val_score
```

```python
lr_model.n_iter_ 
```

```python
train_score = lr_model.score(xtrain,ytrain)
val_score = lr_model.score(xval,yval)
test_score = lr_model.score(xtest,ytest)

train_score, val_score, test_score
```

[4-2] 확률적 경사하강법 기반 학습 <hr>
- 대량의 데이터를 학습 시 랜덤으로 데이터셋 1개 추출해서 학습 진행 => 최적화
- 장점 : 속도가 빨라짐
- 단점 : 정확도 떨어짐
- 개선 : 일정 데이터를 묶어서 학습 후 최적화 진행 => 미니배치 기법  # defalt : 32개 / 몇개로 묶어야 잘나올지 고민 -> 하이퍼 파라미터
- 학습모델 선정 필요

```python
# 모듈 로딩 SGDClassifier => 분류 알고리즘을 SGD방법으로 학습 진행
# from sklearn.metrics import 
from sklearn.linear_model import SGDClassifier
```

```python
# LogisticRegression 알고리즘으로 학습 진행
lr_sgd = SGDClassifier(loss = 'log_loss', random_state=5)
lr_sgd.fit(xtrain,ytrain)
```

```python
# 모델 파라미터 확인
print('n_iter :', lr_sgd.n_iter_)
print(f'[BEFORE] lr_sgd.coef_ :\n{lr_sgd.coef_}')
print(f'[BEFORE] lr_sgd.intercept_ :\n{lr_sgd.intercept_}')
```

```python
lr_sgd.score(xtrain,ytrain), lr_sgd.score(xval,yval),lr_sgd.score(xtest,ytest)
```

```python
# 추가된 데이터에 대한 추가 학습 진행 => 기존 학습으로 최적화 된 상태 유지
import numpy as np

lr_sgd.partial_fit(xval,yval,np.unique(yval))
```
